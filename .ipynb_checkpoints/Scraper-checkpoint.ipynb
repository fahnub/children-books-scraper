```python
from docx import Document
import requests
import PyPDF2
import re
```

```python
file = open('ChildrensBookswithNoCopyright.pdf', 'rb')
reader = PyPDF2.PdfFileReader(file)
```

```python
pages = reader.numPages
```

```python
book_links = []
```

```python
for page in range(1, pages):
    page_info = reader.getPage(page)
    page_text = page_info.extractText()
    page_text = page_text.replace(' ', '')
    urls = re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', page_text)
    book_links.extend(urls)
```

```python
book_links
```

```python
for link in book_links:
    index = book_links.index(link) + 1
    link = link.replace('ebooks', 'files')
    link = link + '/' + link.split('/')[-1] + '.txt'
    print(link)
    
    page = requests.get(link)
    
    document = Document()
    document.add_paragraph(page.text)

    document.save(f'Training Story {index}.docx')
```
